```python
from transformers import AutoModelForCausalLM, AutoTokenizer

checkpoint = "Qwen/Qwen2-1.5B-Instruct"

model = AutoModelForCausalLM.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
def generate (sys, assis, user):
  messages = [
      {"role": "system", "content": sys},
      {"role": "assistant", "content": assis},
      {"role": "user", "content": user}
  ]
  text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
  )
  model_inputs = tokenizer([text], return_tensors="pt")

  generated_ids = model.generate(
    model_inputs.input_ids,
    max_new_tokens=512
  )
  generated_ids = [
      output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
  ]

  response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

  return response
```

```python
#prompt = "Give me a short introduction to Yellow River."
prompt = "Tell me a story about dragon"
system = "You are a story teller."

response = generate(system, "", prompt)
print(response)
```

```python
i=0
while i < 5:
  i = i + 1
  response = generate(system, response, "Please continue")
  print(response)
```
